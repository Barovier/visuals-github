```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# increase the size and quality of any plots we produce
import matplotlib as mpl
from matplotlib import rcParams

mpl.rcParams["figure.dpi"] = 300
```

```python
df = pd.read_csv("Data_Extract_From_Doing_Business/21c884f4-6bee-4f52-a50e-8dda2bf07440_Data.csv")
df['2020 [YR2020]'] = df['2020 [YR2020]'].astype(float)
df['Economy Name'] = df['Economy Name'].astype(str)
```

```python
df = df[df['Series Code'].notna()]
```

```python
meta_df = pd.read_excel("Data_Extract_From_Doing_Business_Metadata.xlsx")
df['Income Group'] = meta_df['Income Group']
```

```python
a = meta_df.groupby('Income Group').size()
a.plot.pie(figsize=(4,4))
plt.title('Relative Size of Each Income Group')
```

```python
df = df.groupby(['Income Group'])['2020 [YR2020]'].mean()
df.plot.bar()
```

```python
df.boxplot('2020 [YR2020]', by='Income Group')
plt.xticks(rotation=45)
rcParams['axes.titlepad'] = 40
```

```python
dates = df.rename(columns={'2004 [YR2004]': 2004, '2005 [YR2005]': 2005, 
                           '2006 [YR2006]': 2006, 
                              '2007 [YR2007]': 2007, '2008 [YR2008]': 2008, 
                           '2009 [YR2009]': 2009, 
                              '2010 [YR2010]': 2010, '2011 [YR2011]': 2011,
                           '2012 [YR2012]': 2012, 
                              '2013 [YR2013]': 2013,'2014 [YR2014]': 2014, 
                           '2015 [YR2015]': 2015,'2016 [YR2016]': 2016, 
                              '2017 [YR2017]': 2017,'2018 [YR2018]': 2018,
                           '2019 [YR2019]': 2019,'2020 [YR2020]': 2020})
melt = dates.melt(id_vars=['Economy Name', 'Series Code', 
                           'Economy Code', 'Category', 'Series Name', 
                           'Income Group'], value_name='score')
```

```python
sns.lineplot(x='variable',y='score',hue='Income Group',data=melt)
```
